## **M√©todos Num√©ricos**
Dado que en muchos casos no podemos obtener una soluci√≥n anal√≠tica (exacta) para la integral del trabajo, utilizamos m√©todos num√©ricos para aproximar el valor del trabajo. Uno de los m√©todos m√°s comunes es la regla del trapecio, que divide el √°rea bajo la curva en peque√±os trapecios y suma sus √°reas.

```python
# Definici√≥n de la funci√≥n de fuerza variable
def fuerza_variable(x):
    return 5 * x

# Definici√≥n de la funci√≥n para calcular el trabajo usando la regla del trapecio
def calcular_trabajo(fuerza, a, b, n):
    x = np.linspace(a, b, n+1)  # Dividimos el intervalo en n subintervalos
    dx = (b - a) / n  # Diferencial de x (dr)
    trabajo = 0.5 * dx * (fuerza(x[0]) + 2 * np.sum(fuerza(x[1:-1])) + fuerza(x[-1]))
    return trabajo
```

```python
a = 0  # L√≠mite inferior de la integral
b = 10  # L√≠mite superior de la integral
n = 100  # N√∫mero de subdivisiones

# Calculo de el trabajo
trabajo = calcular_trabajo(fuerza_variable, a, b, n)
print(f"El trabajo realizado es: {trabajo:.2f} J")
```

```python
x = np.linspace(a, b, 100)
fuerza = fuerza_variable(x)
plt.plot(x, fuerza, label='Fuerza variable')
plt.fill_between(x, fuerza, alpha=0.3)
plt.title('Fuerza Variable vs. Desplazamiento')
plt.xlabel('Desplazamiento (m)')
plt.ylabel('Fuerza (N)')
plt.legend()
plt.grid(True)
plt.show()
```
