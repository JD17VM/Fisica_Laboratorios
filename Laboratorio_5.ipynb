# **Conservación de la Energía Mecánica**


## **Objetivo**

El objetivo del código es demostrar, mediante simulación y visualización gráfica, el principio de conservación de la energía mecánica. Esto se logra calculando y analizando cómo la energía cinética y la energía potencial gravitatoria de un objeto en caída libre varían con el tiempo, mostrando que la energía mecánica total se mantiene constante en ausencia de fuerzas externas como la fricción.

## **Desarrollo**

Se presenta el código en Python para observar la conservación de la energía mecánica en un sistema donde un objeto se mueve verticalmente bajo la influencia de la gravedad.

```python
# Definir parámetros iniciales
m = 1.0  # Masa del objeto (kg)
v0 = 0.0  # Velocidad inicial (m/s)
h0 = 10.0  # Altura inicial (m)
g = 9.81  # Aceleración debido a la gravedad (m/s^2)
t_max = 2.0  # Tiempo máximo de simulación (segundos)
dt = 0.01  # Paso de tiempo (segundos)
```

```python
# Inicializar arrays para el tiempo, energía cinética, energía potencial y energía mecánica
tiempo = np.arange(0, t_max, dt)

# Cálculo de la velocidad y altura durante la caída libre
# Suponiendo que el objeto está cayendo libremente desde una altura inicial h0
velocidad = v0 + g * tiempo  # Velocidad en función del tiempo (v = v0 + g*t)
altura = h0 - 0.5 * g * tiempo**2  # Altura en función del tiempo (h = h0 - 1/2 g t^2)
```

```python
# Evitar que la altura sea negativa
altura[altura < 0] = 0

# Calcular energías
E_K = 0.5 * m * velocidad**2  # Energía cinética
E_P = m * g * altura  # Energía potencial
E_m = E_K + E_P  # Energía mecánica total
```

```python
# Graficar resultados
plt.figure(figsize=(10,6))
plt.plot(tiempo, E_K, label='Energía Cinética (E_K)', color='blue')
plt.plot(tiempo, E_P, label='Energía Potencial (E_P)', color='red')
plt.plot(tiempo, E_m, label='Energía Mecánica Total (E_m)', color='green', linestyle='--')
plt.xlabel('Tiempo (s)')
plt.ylabel('Energía (Joules)')
plt.title('Conservación de la Energía Mecánica')
plt.legend()
plt.grid(True)
plt.show()
```

## **Resultados**

En el experimento se observa que la energía potencial gravitatoria disminuye a medida que el objeto cae, mientras que la energía cinética aumenta debido al incremento de la velocidad. La energía mecánica total permanece constante durante todo el proceso, confirmando el principio de conservación de la energía en ausencia de fuerzas externas. Esto demuestra que la energía se transforma entre sus formas cinética y potencial, pero su suma total no cambia.
