# **Resolución de la Ecuación de la Segunda Ley de Newton**

## **La Segunda Ley de Newton**
La segunda ley de Newton describe cómo se relaciona la aceleración de un objeto con la fuerza neta aplicada sobre él y su masa. Esta ley es expresada matemáticamente de la siguiente forma:

> F=m⋅a

*   F es la fuerza neta aplicada (en Newtons, 𝑁).
*   m es la masa del objeto (en kilogramos, 𝑘𝑔).
*   a es la aceleración del objeto (en metros por segundo al cuadrado, 𝑚/𝑠2).

Esta ecuación expresa que la aceleración de un objeto es directamente proporcional a la fuerza neta aplicada y inversamente proporcional a la masa del objeto. Es decir, a mayor fuerza aplicada, mayor aceleración, pero cuanto mayor sea la masa del objeto, menor será su aceleración para una misma fuerza aplicada.

## **Aplicación Numérica de la Segunda Ley**
En muchos problemas, conoceremos la fuerza aplicada y la masa del objeto, y podremos calcular la aceleración utilizando la segunda ley de Newton. En otros casos, también podremos hacer simulaciones numéricas para calcular la posición y velocidad de un objeto en función del tiempo cuando se le aplica una fuerza.

## **Método Numérico (Ecuación de Movimiento)**
Podemos usar un método numérico para resolver la posición 𝑥(𝑡), la velocidad 𝑣(𝑡), y la aceleración 𝑎(𝑡) de un objeto en función del tiempo, dado que la fuerza aplicada cambia con el tiempo.

La ecuación básica es:

F(t)=m⋅a(t)

Utilizando una integración numérica, podemos resolver estas ecuaciones para obtener la posición y velocidad del objeto a lo largo del tiempo.

```python
# Definir la fuerza variable en función del tiempo
def fuerza_variable(t):
    return 10 * np.sin(t)  # Se puede modificar esta relación según el problema específico

# Definir la función que resuelve la ecuación de la segunda ley de Newton
def segunda_ley_newton(fuerza, m, t):
    # Calcular la aceleración a partir de la fuerza
    a = fuerza(t) / m
    return a
```

```python
m = 2  # Masa del objeto en kg
t = np.linspace(0, 10, 100)  # Tiempo de 0 a 10 segundos

# Calculo la aceleración
a = segunda_ley_newton(fuerza_variable, m, t)
```

```python
plt.plot(t, a, label='Aceleración')
plt.title('Aceleración en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cálculo de la velocidad mediante integración numérica
v = np.cumsum(a * (t[1] - t[0]))  # Integración la aceleración para obtener la velocidad

plt.plot(t, v, label='Velocidad', color='orange')
plt.title('Velocidad en función del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```

## **Explicación del Código**

1. **Fuerza Variable:** La función fuerza_variable(t) define cómo varía la fuerza aplicada al objeto en función del tiempo. En este ejemplo, se ha usado una fuerza sinusoidal, pero puedes cambiar la relación según el problema específico.

2. **Segunda Ley de Newton:** La función segunda_ley_newton utiliza la fórmula de la segunda ley de Newton para calcular la aceleración 𝑎(𝑡) en función de la fuerza aplicada y la masa.

3. **Integración Numérica:** Para obtener la velocidad, integramos la aceleración utilizando el método de suma acumulativa.

4. **Gráficos:** Se generan gráficos para mostrar cómo varían la aceleración y la velocidad del objeto en función del tiempo.
